```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.backends.backend_pdf import PdfPages
import seaborn as sns
data = {'GDP_Quarter': ['Current Dollar Personal Income', 'Disposable Personal Income', 'Personal Saving'],
            'Quarter_three': ['199.5B', '95.8 B', '776.9B'],
            'Quarter_four': ['230.2B', '190.4B', '815.5 B'],
            'Quarter_one': ['404.4B', '266.7B', '796.6B'],
            'Quarter_two': ['315.7B', '260.4B', '1.13T']}

gdp_dataframe = pd.DataFrame(data)
```

```python
gdp_dataframe
```

```python
gdp_dataframe['Quarter_three'] = [199.5 * 1e9, 95.8 * 1e9, 776.9 * 1e9]

plt.figure(figsize=(5, 6))
plt.bar(gdp_dataframe['GDP_Quarter'], gdp_dataframe['Quarter_three'], color='green')
plt.xlabel("Personal Income Category")
plt.ylabel("Quarter Three (in Billions USD)")
plt.title("Quarter Three Personal Income Data")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
